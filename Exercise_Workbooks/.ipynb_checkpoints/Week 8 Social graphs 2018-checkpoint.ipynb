```python
import pandas as pd
import networkx as nx
import numpy as np
import re
import matplotlib.pyplot as plt
from fa2 import ForceAtlas2
import json
from scipy import stats
import math
from scipy.special import zeta
```

```python
H115 = pd.read_csv("H115_tw.csv")
```

```python
authors = H115["tw_name"].unique()
```

```python
G = nx.DiGraph()
for i in H115.index:
    G.add_node(H115.loc[i].at["tw_name"], party = H115.loc[i].at["Party"])
for author in authors:
    with open("../tweets/"+author) as f:
        string = str(f.readlines())
    matches = re.findall(r'RT @(\w*)', string)
    for match in matches:
        if author==match:
            continue
        if match not in authors:
            continue
        if G.has_edge(author, match):
            G[author][match]["weight"] +=1
        else:
            G.add_edge(author,match,weight=1)
```

```python
G.number_of_edges()
```

```python
forceatlas2 = ForceAtlas2(
                          # Behavior alternatives
                          outboundAttractionDistribution=False,  # Dissuade hubs
                          linLogMode=False,  # NOT IMPLEMENTED
                          adjustSizes=False,  # Prevent overlap (NOT IMPLEMENTED)
                          edgeWeightInfluence=1.0,

                          # Performance
                          jitterTolerance=1.0,  # Tolerance
                          barnesHutOptimize=True,
                          barnesHutTheta=1.2,
                          multiThreaded=False,  # NOT IMPLEMENTED

                          # Tuning
                          scalingRatio=0.01,
                          strongGravityMode=False,
                          gravity=10.0,

                          # Log
                          verbose=True)
positions = forceatlas2.forceatlas2_networkx_layout(nx.to_undirected(G), pos=None, iterations=4000)
```

```python
fig = plt.figure(figsize=(20,14)) 
dicts = dict(G.nodes())
colors = ["#4091db" if dicts[item]["party"]=="Democratic" else "#992D2D" for item in list(G.nodes())]
sizes = [10*dict(G.degree())[rep] for rep in list(G.nodes())]
edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())
nx.draw(G, positions, node_size=sizes,edgelist=edges, edge_color=np.log10([weight+1 for weight in weights]), node_color = colors, with_labels=False, width=1,edge_cmap=plt.cm.Greys)
plt.show()
```

```python
summedweightsRep = []
summedweightsDem = []
parties = dict(G.nodes(data='party'))
for author in authors:
    newedges = list(G.out_edges(author))
    summedweight = sum([G.get_edge_data(*edge).values()[0] for edge in newedges])
    if parties[author] == "Republican":
        summedweightsRep.append(summedweight)
    else:
        summedweightsDem.append(summedweight)
```

```python
fig = plt.figure(figsize=(20,14))
itera = list(range(min(summedweightsRep)-1,max(summedweightsRep)+1))
newitera = [x+1 for x in itera]
hist = np.histogram(summedweightsRep,bins=newitera)
plt.bar(hist[1][:-1],hist[0],align='center',width=0.3,color="r",label="Republicans")
itera = list(range(min(summedweightsDem)-1,max(summedweightsDem)+1))
newitera = [x+1.5 for x in itera]
hist = np.histogram(summedweightsDem,bins=newitera)
plt.bar(hist[1][:-1],hist[0],align='center',width=0.3,color="b",label="Democrats")
plt.legend()
plt.show()
```

```python
summedweightsReptoDem = []
summedweightsDemtoRep = []
parties = dict(G.nodes(data='party'))
for author in authors:
    newedges = list(G.out_edges(author))
    if parties[author] == "Republican":
        summedweight = sum([G.get_edge_data(*edge).values()[0] for edge in newedges if parties[edge[1]]=="Democratic"])
        summedweightsReptoDem.append((author,summedweight))
    else:
        summedweight = sum([G.get_edge_data(*edge).values()[0] for edge in newedges if parties[edge[1]]=="Republican"])
        summedweightsDemtoRep.append((author,summedweight))
```
